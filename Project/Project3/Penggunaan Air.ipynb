```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./data/water_usage.csv')
```

```python
# Create figure
plt.figure(figsize=(8, 8), dpi=300)
# Create pie plot
plt.pie('Percentage', explode=(0, 0, 0.1, 0, 0, 0), labels='Usage', data=data, autopct='%.0f%%')
# Add title
plt.title('Water usage')
# Show plot
plt.show()
```
